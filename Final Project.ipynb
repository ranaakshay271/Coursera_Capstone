```python
import geocoder
def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Delhi, India'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(neighborhood) for neighborhood in dl_df["Neighborhood"].tolist() ]
```

```python
dl_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

dl_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

fixed_columns = [dl_onehot.columns[-1]] + list(dl_onehot.columns[:-1])
dl_onehot = dl_onehot[fixed_columns]

print(dl_onehot.shape)
dl_onehot.head()
```

```python
dl_grouped = dl_onehot.groupby(["Neighborhoods"]).mean().reset_index()

print(dl_grouped.shape)
dl_grouped
```

```python
len(dl_grouped[dl_grouped["Indian Restaurant"] > 0])
```

```python
dl_ir = dl_grouped[["Neighborhoods","Indian Restaurant"]]
```

```python
dl_ir.head()
```

```python
kclusters = 3

dl_clustering = dl_ir.drop(["Neighborhoods"], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dl_clustering)

kmeans.labels_[0:10]
```

```python
dl_merged = dl_ir.copy()

dl_merged["Cluster Labels"] = kmeans.labels_
```

```python
dl_merged.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
dl_merged.head()
```

```python
dl_merged = dl_merged.join(dl_df.set_index("Neighborhood"), on="Neighborhood")

print(dl_merged.shape)
dl_merged.head()
```

```python
print(dl_merged.shape)
dl_merged.sort_values(["Cluster Labels"], inplace=True)
dl_merged
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(dl_merged['Latitude'], dl_merged['Longitude'], dl_merged['Neighborhood'], dl_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

This represents a great opportunity and high potential areas to open new indian restaurants as there is very little to no competition from existing indian restaurants. Meanwhile, indian restaurants in cluster 0 are likely suffering from intense competition due to oversupply and high concentration of indian restaurants. Therefore, this project recommends to capitalize on these findings to open new indian restaurants in neighborhoods in cluster 1 with little to no competition. People with unique selling propositions to stand out from the competition can also open new indian restaurants in neighborhoods in cluster 2 with moderate competition. Lastly, people are advised to avoid neighborhoods in cluster 0 which already have high concentration of indian restaurants and suffering from intense competition.
